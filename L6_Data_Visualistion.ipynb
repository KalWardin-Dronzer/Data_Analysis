### Question 2: How does the education level vary with income in the dataset?

```python
X.head()
```

```python
X['education'].value_counts()
```

```python
X['income'].value_counts()
```

```python
plt.bar(X['education'], X['income'])
plt.xticks(rotation = 90)
# hence neither scatter nor bar(until it's stacked) is suitable for interpretation
#of categorical features
```

```python
""" .unstack() - used to pivot a level of the (necessarily hierarchical) index labels.
 used to unstack the values when values are stakced on each other.
# normalise = True -
multi line comment only works if written above the code else it will print the comment
since it is a string that's not stored in any var"""

edu_income = X.groupby('education')['income'].value_counts().unstack()
edu_income
```

```python
edu_income.plot(kind = 'bar', stacked = True)
plt.xticks(rotation = 90)
```

```python
edu_income.plot(kind = 'line', stacked = True)
plt.xticks(rotation = 90)
#while stacking line chart beware because it shows aggregate value.
# so it's better to keep stacked = False
```

```python
# plots line chart
plt.plot(edu_income.index, edu_income)
plt.xticks(rotation = 90)
# shows more x labels but not label of the line
```

```python
edu_income.plot(kind = 'line', stacked = False)
plt.xticks(rotation = 90)
# in cases like these it's better to use line chart rather than bar chart
```

```python
pd.pivot_table(X, index = 'education', columns = 'income', values = 'age', aggfunc = 'count')
```

```python
edu_income = X.groupby('education')['income'].value_counts(normalize = True).unstack()
edu_income
# gives value in proportion i.e., if 0.627 it means 62.7%
```

#### Question 4: How is the marital status distributed among different occupations?

```python
X.head()
```

```python
moccu = X.groupby('marital-status')['occupation'].value_counts().unstack()
moccu
```

```python
moccu.plot(kind = 'bar', stacked = True)
```

```python
mooc = pd.pivot_table(X, index = 'occupation', columns = 'marital-status', values = 'age', aggfunc = 'count')
mooc
```

```python
# gives better result than the previous one
mooc.plot(kind = 'bar', stacked = True)
plt.xlabel('occupation')
plt.title('relation b/w occ & M-staus')
plt.ylabel('m-status')
plt.legend(title = 'mooc')
plt.show()
plt.figure(figsize = (15, 10))
```
